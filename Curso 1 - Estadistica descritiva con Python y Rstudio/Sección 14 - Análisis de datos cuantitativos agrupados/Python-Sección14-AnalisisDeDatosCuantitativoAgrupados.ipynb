```python
#Haciendo divisiones en la edad
age = pd.cut(titanic['age'], #Escojo como estudio la edad
            bins = [0,18,25,35,50,65,100],#Separo por edad de 0 hasta 18, 18 hasta 25 y asi....
            include_lowest = True,#Para incluir el ultimo valor en el intervalo, en este caso el lado izquierdo
             #Lado izquierdo sin incluir, lado derecho incluido, con incluide lowest incluye el lado izquierdo
             
            precision = 1, #Precisi√≥n es de 1 unidad. para crear los intervalos en valores con menos decimales
            right = False) #Para que este cerrado por la izquiera y abierto por la derecha-Deve estar en False
age
```

#### Tabla de frecuencias relativas 1 donde el calculo de los intervalos es realizado por los 4 clases vistas anteriormente. escojer solo 1

```python
#Tabla de frecuencias 1
import pandas as pd
import numpy as np

def TablaFrec(x,k,A,p):
    
    L = list(np.min(x)-p/2+A*np.arange(k+1)) #Limites

#Lista de intervalos    
    x_cut = pd.cut(x,
                  bins = L,
                  include_lowest = True,
                  right = False) 
    
    intervals = x_cut.cat.categories #Intervalos unicos
    
    mc = (L[0]+L[1])/2+A*np.arange(k) #Marca de clase
    
#Frecuencia absoluta   
    fr_abs = pd.crosstab(x_cut, 
                         columns = 0,
                         dropna = False) #Frecuencia Absoluta
    fr_rel = round(fr_abs/fr_abs.sum(),2) #Frecuencia relativa

#Frecuencia absoluta acumulada
    a = []
    b = 0
    for c in fr_abs[0]:
        b = c+b
        a.append(b)
    fr_cum_abs = a
    
#Frecuencia relativa acumulada
    a = []
    b = 0
    for c in fr_rel[0]:
        b = c+b
        a.append(b)
    fr_cum_rel = a
    
#Creaci√≥n de la tabla final

    tabla = pd.DataFrame(intervals,
                    columns = ['intervalos'])
    tabla['mc'] = mc
    tabla['fr_abs'] = fr_abs.values.tolist()
    tabla['fr_cum_abs'] = fr_cum_abs
    tabla['fr_rel'] = fr_rel.values.tolist()
    tabla['fr_cum_rel'] = fr_cum_rel
    return tabla
    
```

```python
import seaborn as sb #Para llamar el data frame de titanic 
titanic = sb.load_dataset('titanic')
x = titanic['age']
TablaFrec(x, 6, 1, 0.1) #Pruebo la tabla de frecuencias relativas 
```

```python
titanic = sb.load_dataset('titanic')
data = titanic['age'].dropna()
data = np.array(data)
print('1. Establecemos el numero de clases, utilizando la regla de Sturges:')
k = np.ceil(1 + np.log2(len(data)))
k
print('2. Definimos la amplitud de los intervalos, en este caso seria:')
min = np.min(data)
max = np.max(data)
A = np.ceil((max - min)/k)
A
def freq_table(x, k, A, p):
    L = np.min(x)-p/2+A*np.arange(0,k)
    x_cut = pd.cut(x, bins = L, right=False)
    mc = (L[0]+L[1])/2+A*np.arange(0,(k-1))
    unique,count = np.unique(x_cut, return_counts = True)
        
    abs_freq = count
    acum_abs_freq = np.cumsum(abs_freq)
    rel_freq = count/count.sum()
    acum_rel_freq = np.cumsum(rel_freq)
        
    freq_tab = pd.DataFrame({
                            "value": unique,
                            "mc": mc,
                            "abs_freq": abs_freq,
                            "acum_abs_freq": acum_abs_freq,
                            "rel_freq": rel_freq,
                            "acum_rel_freq": acum_rel_freq,
                            })
    return freq_tab
     
freq_table(data,k,A,0.01)
```
